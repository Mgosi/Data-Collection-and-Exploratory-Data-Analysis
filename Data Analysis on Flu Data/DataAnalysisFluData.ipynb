## Chart 1

```r
plot1<-read.csv("1-Influenza national summary.csv") #read data
w<-c(plot1$Week)
values <-matrix(c(plot1$Total.B ,plot1$Total.A),nrow = 2, ncol =31,byrow = TRUE) #prepare data for stacking barplot
colors=c("green","yellow") #array of color stack

barplot(values,names.arg=w, xlab="Week",ylab="Number of Positive Specimens",col=colors,main="Influenza Positive Tests Reported to CDC by U.S Clinical \n Laboratories,National Summary, 2018-2019 Season",border="black",cex.names = 0.7,cex.axis=0.8,ylim=c(0,14000))

#plotting with 3rd axis
par(new=TRUE)
plot(plot1$X..Positive,type = "l",col="black",xaxt = "n",yaxt= "n",ylab="",xlab = "",ylim=c(5,35))
axis(side=4)
mtext("Percent positives",side = 4, line = 0)
#plotting lines
lines(plot1$Percent.Positive.A,type = "l",col="red",lty=2)
lines(plot1$Percent.Positive.B,type = "l",col="blue",lty=2)

legend("topright", legend= c("B","A", "Percent Positive", "%A", "%B"), col = c("green","yellow","black", "red", "blue"),pch=c(15, 15, NA, NA, NA), inset=0.02,lty=c(0,0,1,2,2), cex = 0.75)
```

## Chart 4

```r
pediatric_deaths <- read.csv("4-PediatricDeaths.csv")

val <- matrix(c(pediatric_deaths$PREVIOUS.WEEK.DEATHS))
s1 <- pediatric_deaths[c(0:52), c(0:4)]
s2 <- pediatric_deaths[c(53:104), c(0:4)]
s3 <- pediatric_deaths[c(105:156), c(0:4)]
s4 <- pediatric_deaths[c(157:208), c(0:4)]

count1 <- sum(s1$NO..OF.DEATHS)
count2 <- sum(s2$NO..OF.DEATHS)
count3 <- sum(s3$NO..OF.DEATHS)
count4 <- sum(s4$NO..OF.DEATHS)
barplot(val, 
        lwd = 3, pch = ".",  
        main="Number of Influenza-Associated Pediatric Deaths by Week of Death", 
        width = 5, ylab= "No. of Deaths", ylim = c(0,30), xlab = "Week of Death", 
        names.arg = pediatric_deaths$WEEK.NUMBER, col = c("darkgreen"),cex.names = 0.75, 
        cex.axis = 0.75, las = 1, beside = T)
text(x=130,y=23,labels = paste("2015-2016 \n No. of Deaths = ", count1), cex = 0.70, font = 2)
text(x=370,y=23,labels = paste("2016-2017 \n No. of Deaths = ", count2), cex = 0.70,font = 2)
text(x=600,y=23,labels = paste("2017-2018 \n No. of Deaths = ", count3), cex = 0.70,font = 2)
text(x=850,y=23,labels = paste("2018-2019 \n No. of Deaths = ", count4), cex = 0.70,font = 2)
legend("topright", title = "Week of Death", border = "black",legend= c("Previous Week"), col = c("darkgreen"),pch=c(15), lty=c(0), cex = 0.85)
```

```r
df<-data.frame(read.csv("5-40to4-StateDatabyWeekforMap_2018-19week40-4.csv"))

colnames(df)[colnames(df)=="STATENAME"] <- "state"

df$level <- as.numeric( substr( df$ACTIVITY.LEVEL, 6, 8 ) )

library(usmap)
library(ggplot2)

plot_usmap(data = df, values = "level", lines = "black") + 
  scale_fill_gradient(name = "ILI Activity Level", low = "greenyellow",
  high = "red", label = scales::comma) + labs(title = "2018-2019 Influenza season from week 40 of 2018 to week 4 of 2019")+
  theme(legend.position = "right")
```
